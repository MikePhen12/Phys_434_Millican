Upon research starting at https://en.wikipedia.org/wiki/Exponential_distribution, I have found that the summation of two identical exponential PDFs is resultant in an Erlang distribution with shape=2.0 and parameter equal to the rate, $\lambda$, in the exponential distribution which happens to be a special case of a gamma distribution. Found at https://reference.wolfram.com/language/ref/GammaDistribution.html \
$ \lambda e^{{-\lambda x}} \rightarrow \lambda ^{2}xe^{-\lambda x} $ \
The sum of more identical exponential distributions also is a gamma distribution with integer "shape" parameters equal to the number of convolutions. These are also all of the Elrang distributions.
